```python
import pandas as pd
from pandas_ods_reader import read_ods
import xgboost as xg
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'datasets/cool_temp_maize_soybean/'
file_names = ['flux_soc_out', 'atmosphere', 'n_out', 'n2o_nh3_out', 'p_out', 'soil_c_out', 'soil_water_out', 'surf_water_', 'temp_out']
start_year = 1998
growth_period = 6
```

```python
def load_files(start_year: str, period: int, file_name : str, path: str) -> pd.DataFrame:
    files = []
    for i in range(period + 1):
        df = read_ods(path + str(start_year) + file_name + '.ods')
        df.set_index('DATE')
        df = df.drop('DOY', axis=1)
        files.append(df)
        start_year += 1
        
    return pd.concat(files)
```

```python
df = load_files(start_year, growth_period, file_names[0], path)
```

```python
X, y = df.drop('CO2_FLUX', axis=1), df.iloc[:, 5]
```

```python
X
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y,
                      test_size = 0.3, random_state = 123)
```

```python
train_y
```

```python
# Instantiation
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 123, booster='gblinear')
```

```python
  
# Fitting the model
xgb_r.fit(train_X, train_y)
```

```python
# Predict the model
pred = xgb_r.predict(test_X)
  
```

```python
# RMSE Computation
rmse = np.sqrt(MSE(test_y, pred))
print("RMSE : % f" %(rmse))
plt.scatter(test_y, pred)
```
